## Removing HTML tags

```python
import re
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
df['text'] = df['text'].apply(remove_html_tags)
```

```python
df['text'].head(10)
```

## Removing URLS

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
df['text'] = df['text'].apply(remove_urls)
```

```python
df['text'].head(10)
```

## Removing Punctuations

```python
import string
def remove_punc(text):
    return text.translate(str.maketrans('', '', string.punctuation))
df['text'] = df['text'].apply(remove_punc)
```

```python
df['text'].head(10)
```

## LowerCasing

```python
df['text'] = df['text'].str.lower()
```

## Chat word treatment

```python
chat_word = {
    "U": "you",
    "R": "are",
    "UR": "your",
    "BRB": "be right back",
    "LOL": "laughing out loud",
    "OMG": "oh my god",
    "IDK": "I don’t know",
    "TBH": "to be honest",
    "BTW": "by the way",
    "GTG": "got to go",
    "BFF": "best friends forever",
    "ROFL": "rolling on the floor laughing",
    "TTYL": "talk to you later",
    "IMO": "in my opinion",
    "IMHO": "in my humble opinion",
    "FYI": "for your information",
    "LMAO": "laughing my ass off",
    "NP": "no problem",
    "THX": "thanks",
    "TY": "thank you",
    "PLZ": "please",
    "CUL8R": "see you later",
    "GR8": "great",
    "NVM": "never mind",
    "SMH": "shaking my head",
    "IDC": "I don’t care",
    "ILY": "I love you",
    "YOLO": "you only live once",
    "TMI": "too much information",
    "AFK": "away from keyboard",
    "OMW": "on my way",
    "FAQ": "frequently asked questions",
    "ASAP": "as soon as possible",
    "FOMO": "fear of missing out"}
def chat_word_conversion(text):
    new_text = []
    for w in text.split(): # it will split the words in a sentence by spaces into a list.
        if w.upper() in chat_word: #Check if the word is in dictionary
            new_text.append(chat_word[w.upper()]) #This finds the full form of the chat word and appends it.
        else:
            new_text.append(w)
    return " ".join(new_text) #joins all words with spaces back into a sentence.  
```

```python
df['text'] = df['text'].apply(chat_word_conversion)
```

```python
df['text'].head(10)
```

## Spelling Correction

```python
from textblob import TextBlob
def Spell_Correct(incorrect_text):
   
    textBlb = TextBlob(str(incorrect_text)) #creating TextBlob object
    text= str(textBlb.correct())
    return text
```

```python
df['text'] = df['text'].apply(Spell_Correct)
```

```python
df['text'].head(10)
```

## Removing emojis

```python
import emoji
def emoji_removal(text):
    text = emoji.demojize(text)
    return text
```

```python
df['text'] = df['text'].apply(emoji_removal) # ye emoji k meaning se replace kre ga
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def token(text):
    doc = nlp(text)
    return [token.text for token in doc]
```

```python
df['text'] = df['text'].apply(token)
```

## Stemming

```python
from nltk import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    return [ps.stem(word) for word in text]
```

```python
df['text'] = df['text'].apply(stemming)
```

```python
df['text'].head(10)
```
